```python
import numpy as np 
import matplotlib.pyplot as plt
import pandas as pd

plt.rcParams['font.family'] = 'Century Schoolbook'

x0 = 100
sigma = 0.6  
t = np.arange(0, 10, 0.01)  
gamma_dic = np.arange(0.2, 1, 0.2)  
gamma_int = np.arange(1, 4, 1)  
gamma = np.append(gamma_dic, gamma_int)  

epsilon_val = []
v_val = []
T05_val = []
law_val = []

plt.figure(figsize=(10, 6))

for gamma_val in gamma:
    x = x0 * np.exp((gamma_val - sigma) * t) 
    
    line, = plt.plot(t, x, label=r'$\gamma = {:.1f}$'.format(gamma_val))
    curve_color = line.get_color()
    
    if gamma_val != sigma:  
        epsilon = gamma_val - sigma
        v = epsilon*x0
        T05 = np.log(2) / abs(epsilon)
        law = f'x(t) = {x0}*e^{round(gamma_val - sigma, 1)}*t' 
        
        epsilon_val.append(round(epsilon, 1))
        v_val.append(round(v, 1)) 
        T05_val.append(round(T05, 1))
        law_val.append(law)

        plt.vlines(T05, 0, x0 * np.exp((gamma_val - sigma) * T05), color=curve_color, linestyle='--')
        plt.plot(T05, x0 * np.exp((gamma_val - sigma) * T05), 'o', color=curve_color)


plt.xlim(0, 8)
plt.ylim(0, 1000)
plt.xlabel('Час, t', fontsize=16)
plt.ylabel('Кількість осіб, x', fontsize=16)
plt.title(r'Графік зміни чисельності популяції, $x = x_0 \cdot e^{(\gamma - \sigma) \cdot t}$', fontsize = 18, fontweight='bold')
plt.text(3.2, 900, r'$\sigma = 0.6$; $x_0 = 100$', fontsize = 16)
plt.legend(fontsize=14, loc='upper right')
plt.grid()
plt.show()

data = {
    'Коефіцієнт розмноження, γ': gamma,
    'Коефіцієнт росту, ε': epsilon_val,
    'Швиткість зміни чисельності популяції': v_val,
    'Час T0.5': T05_val,
    'Закон, що описує зміну чисельності популяції': law_val}

df = pd.DataFrame(data)

df
```

```python
x0 = 500
gamma = 0.4  
t = np.arange(0, 10, 0.01)  
sigma_dic = np.arange(0.2, 1, 0.2)  
sigma_int = np.arange(1, 4, 1)  
sigma = np.append(sigma_dic, sigma_int)  

epsilon_val = []
v_val = []
T05_val = [] 
law_val = []

plt.figure(figsize=(10, 6))

for sigma_val in sigma:
    x = x0 * np.exp((gamma - sigma_val) * t) 
    
    line, = plt.plot(t, x, label=r'$\sigma = {:.1f}$'.format(sigma_val))
    curve_color = line.get_color()
    
    if sigma_val != gamma:  
        epsilon = gamma - sigma_val
        v = epsilon*x0
        T05 = np.log(2) / abs(epsilon) 
        law = f'x(t) = {x0}*e^{round(gamma - sigma_val, 1)}*t'
        
        epsilon_val.append(round(epsilon, 1))
        v_val.append(round(v))
        T05_val.append(round(T05, 1))
        law_val.append(law)


        plt.vlines(T05, -100, x0 * np.exp((gamma - sigma_val) * T05), color=curve_color, linestyle='--')
        plt.plot(T05, x0 * np.exp((gamma - sigma_val) * T05), 'o', color=curve_color)  
    else:
        nan = np.nan 
        epsilon_val.append(nan)
        v_val.append(nan)
        T05_val.append(nan)
        law_val.append(nan)

plt.xlim(0, 4)
plt.ylim(-100, 1200)
plt.xlabel('Час, t', fontsize=16)
plt.ylabel('Кількість осіб, x', fontsize=16)
plt.title(r'Графік зміни чисельності популяції, $x = x_0 \cdot e^{(\gamma - \sigma) \cdot t}$', fontsize = 18, fontweight='bold')
plt.text(1.6, 1100, r'$\gamma = 0.4$; $x_0 = 100$', fontsize = 16)
plt.legend(fontsize=13, loc='upper left')
plt.grid()
plt.show()

data = {
    'Коефіцієнт смертності, σ': sigma,
    'Коефіцієнт росту, ε': epsilon_val,
    'Швиткість зміни чисельності популяції': v_val,
    'Час T0.5': T05_val,
    'Закон, що описує зміну чисельності популяції': law_val
    }

df = pd.DataFrame(data)

df
```

```python
gamma = 1.4   
sigma = 0.4  
t = np.arange(0, 10, 0.01)  
x0 = np.geomspace(10, 1000, num=6)

x0_values = []
v_val = []
T05_val = [] 
law_val = []

plt.figure(figsize=(10, 6))

for x0_val in x0:
    x = int(x0_val) * np.exp((gamma - sigma) * t) 
    
    line, = plt.plot(t, x, label=r'$x_0 = {:.0f}$'.format(x0_val))
    curve_color = line.get_color()
    
    if sigma != gamma:  
        epsilon = gamma - sigma
        v = epsilon*x0_val
        T05 = np.log(2) / abs(epsilon) 
        law = f'x(t) = {x0_val.round(1)}*e^({round(gamma - sigma, 1)}*t)'
        
        x0_values.append(round(x0_val))
        v_val.append(round(v))
        T05_val.append(round(T05, 1))
        law_val.append(law)

        plt.vlines(T05, -100, x0_val * np.exp((gamma - sigma) * T05), color=curve_color, linestyle='--')
        plt.plot(T05, x0_val * np.exp((gamma - sigma) * T05), 'o', color=curve_color)  

plt.xlim(0, 4)
plt.ylim(-100, 1200)
plt.xlabel('Час, t', fontsize=16)
plt.ylabel('Кількість осіб, x', fontsize=16)
plt.title(r'Графік зміни чисельності популяції, $x = x_0 \cdot e^{\epsilon \cdot t}$', fontsize = 18, fontweight='bold')
plt.text(1.25, 1100, r'$\epsilon = const$; варіація: $x_0$ ', fontsize = 16)
plt.legend(fontsize=13, loc='upper right')
plt.grid()
plt.show()

data = {
    'Початкова чисельність популяції': x0,
    'Швидкість зміни чисельності популяції': v_val,
    'Час T0.5': T05_val,
    'Закон, що описує зміну чисельності популяції': law_val
}
df = pd.DataFrame(data)

df
```
